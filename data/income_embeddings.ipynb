```python
dim_k = 200
```

```python
partition = Z[:dim_k]
```

```python
kmeans = faiss.Kmeans(partition.shape[1], 10, niter=20, nredo=1, verbose=False,
                                          min_points_per_centroid=5 + 15, gpu=1)
```

```python
kmeans.train(partition)
```

```python
D, I = kmeans.index.search(partition, 1)
```
